#### Categorical Feature

```python
# Let's capture nan value first 
# We will take categorical feature first

categorical_feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes == 'O']
categorical_feature_nan
```

```python
for feature in categorical_feature_nan:
    print('{} has {}% missing values'.format(feature,np.round(dataset[feature].isnull().mean()*100,4)))
```

```python
dataset
```

```python
## whenever you have missing value in categorical feature try to replace it with other label, it acts like a new category

def replace_cat_feature_nan(dataset,feature):
    data=dataset.copy()
    data[feature]=data[feature].fillna('Missing')
    return data

dataset = replace_cat_feature_nan(dataset,categorical_feature_nan)

dataset[categorical_feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes== 'O']
categorical_features
```

#### Numerical Feature

```python
# Let's capture nan value first 
# Now We will take numerical feature first

numerical_feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes!= 'O']
numerical_feature_nan

for feature in numerical_feature_nan:
    print('{} has {}% missing values'.format(feature,np.round(dataset[feature].isnull().mean()*100,4)))
```

```python
### In EDA, we have see that numerical features have so many outliers. So if you  have many outliers go
### with the approach of replacing missing value with median or mode

for feature in numerical_feature_nan:
    median_value = dataset[feature].median()
    
    # creating a new feature to capture nan values
#     dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)

dataset[numerical_feature_nan].isnull().sum()
```

### Handling Temporarl Variables

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature] = dataset['YrSold']-dataset[feature]
```

### Converting Skewed Distribution To Normal Distribution

```python
### We need to leave handle only those value who don't have 0 in there unique() value list

num_feature_without_0 = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in num_feature_without_0:
    dataset[feature]= np.log(dataset[feature])
```

```python
dataset.head()
```

### Handling Rare Categorical Feature
We will assign a new label to categorical variable which are less than 1% of the observation

```python
for feature in categorical_feature:
    temp_count=dataset.groupby(feature)['SalePrice'].count()
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```
